```python
import pandas as pd
```

Data

```python
flight_itineraries_data = pd.read_csv('../../data/flight_itineraries_data.csv')

display(flight_itineraries_data.shape)
display(flight_itineraries_data.head())
```

```python
cost_of_hour = 20
cost_of_stop = 50

flight_itineraries_data['flight_score'] = flight_itineraries_data['price_amount'] + flight_itineraries_data['total_flight_duration'].divide(60).multiply(cost_of_hour) + flight_itineraries_data['segment_count'].multiply(cost_of_stop)

flight_itineraries_data['flight_score'] = flight_itineraries_data['flight_score'].round(2)

flight_itineraries_data.head()
```

```python
# group and get lowest score val
min_flight_score_indices = flight_itineraries_data.groupby(['departure_query', 'arrival_query'])['flight_score'].idxmin()

# filter
filtered_flight_itineraries = flight_itineraries_data.loc[min_flight_score_indices]
filtered_flight_itineraries = filtered_flight_itineraries.reset_index(drop=True)

display(filtered_flight_itineraries.shape)
display(filtered_flight_itineraries.head())
```
